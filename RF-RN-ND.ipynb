Roger Federer, Rafael Nadal and Novak Djokovic
===

Ranking of Roger Federer, Rafael Nadal, and Novak Djokovic. There would be a smart way to scrape it straight from the ATP site, but for now, it reads it from the manually updated text-files.

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 150
```

## Load Data

```python
# Roger Federer
data = np.loadtxt('data/RF-RN-ND/RF.txt', dtype='int')

ndat = np.size(data, 0)
dates = np.zeros(ndat)
for i in np.arange(ndat):
    dates[i] = datetime.date(data[i,0], data[i,1], data[i,2]).toordinal()

RF = np.array([dates, data[:,3]])

# Rafael Nadal
data = np.loadtxt('data/RF-RN-ND/RN.txt', dtype='int')

ndat = np.size(data, 0)
dates = np.zeros(ndat)
for i in np.arange(ndat):
    dates[i] = datetime.date(data[i,0], data[i,1], data[i,2]).toordinal()

RN = np.array([dates, data[:,3]])

# Novak Djokovic
data = np.loadtxt('data/RF-RN-ND/ND.txt', dtype='int')

ndat = np.size(data, 0)
dates = np.zeros(ndat)
for i in np.arange(ndat):
    dates[i] = datetime.date(data[i,0], data[i,1], data[i,2]).toordinal()

ND = np.array([dates, data[:,3]])

# Labels
xticks = np.array([datetime.date(x,1,1).toordinal() for x in np.arange(1998, data[0, 0]+1)])
xticklabels = [str(x) for x in np.arange(1998, data[0, 0]+1)]
```

## Plots

```python
fig1 = plt.figure()
ax = plt.gca()

# Plot the data
plt.semilogy(RF[0,:], RF[1,:], label='Roger Federer', zorder=10)
plt.semilogy(RN[0,:], RN[1,:], label='Rafael Nadal', zorder=9)
plt.semilogy(ND[0,:], ND[1,:], label='Novak Djokovic', zorder=8)

# Title, labels, legend
plt.title('Ranking History')
plt.legend(loc=4)
plt.ylim([1100, 0.9])
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=45)
ax.set_yticklabels(['.01', '.1', '1', '10', '100', '1000'])

# Save and show
plt.savefig('plots/RF-RN-ND_log.png', bbox_inches='tight')
plt.show()
```

```python
fig2 = plt.figure()
ax = plt.gca()

# Plot the data
plt.plot(RF[0, :], RF[1, :], 'C0', lw=1, ms=4, label='Roger Federer', zorder=10)
plt.plot(RN[0, :], RN[1, :], 'C1', lw=1, ms=4, label='Rafael Nadal', zorder=9)
plt.plot(ND[0, :], ND[1, :], 'C2', lw=1, ms=4, label='Novak Djokovic', zorder=8)

# Title, labels, legend
plt.title('Ranking History')
plt.legend(loc=8)
plt.ylim([10.5, .5])
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=45)
ax.set_yticks(np.arange(10)+1)

# Start with 2002
plt.xlim([730851, ax.get_xlim()[1]])

# Save and show
plt.savefig('plots/RF-RN-ND_lin.png', bbox_inches='tight')
plt.show()
```
